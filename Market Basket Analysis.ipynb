```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

dataset = [["Bread", "Milk", "Beer"],
           ["Bread", "Diapers", "Eggs"],
           ["Milk", "Diapers", "Beer", "Cola"],
           ["Bread", "Milk", "Diapers", "Beer"],
           ["Bread", "Milk", "Cola"]]


```

```python
te = TransactionEncoder()
te = te.fit(dataset)             # Read all unique items from the dataset
te_ary = te.transform(dataset)   # Converts array into hot coded matrix that is it contains 1 and 0
print(te.columns_)               # Prints all the unique items of the dataset which are columns to the hot coded matrix

#As the columns are string so the columns are in alphabetically ordered
print("\n")
print(te_ary)                    # Prints the hot coded matrix which is the required input format for apriori

df = pd.DataFrame(te_ary, columns = te.columns_) # It converts the hot coded matrix to a Pandas DataFrame

# Finds the frequency of the item
frequent_itemsets = apriori(df, min_support = 0.3, use_colnames = True) # min_support means show the data that has frequency > 60 %

print("\n")
print(frequent_itemsets)

from mlxtend.frequent_patterns import association_rules
asso_rule_table = association_rules(frequent_itemsets, metric = "support", min_threshold = 0.4)

# Here it gives another constraint to the "support"(metric) column that the value should be greater than "0.5"(min_threshold)

asso_rule_table # Printing the association_rule table
```

```python
asso_rule_table['length_consequent'] = asso_rule_table['consequents'].apply(lambda x : len(list(x)))
asso_rule_table
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

data =    [["Bread", "Milk", "Beer"],
           ["Bread", "Diapers", "Eggs"],
           ["Milk", "Diapers", "Beer", "Cola"],
           ["Bread", "Milk", "Diapers", "Beer"],
           ["Bread", "Milk", "Cola"]]

te = TransactionEncoder()
te = te.fit(data) 
te_ary = te.transform(data)

df = pd.DataFrame(te_ary, columns = te.columns_)

frequent_itemsets = apriori(df, min_support = 0.3, use_colnames = True) 

from mlxtend.frequent_patterns import association_rules
rules_table = association_rules(frequent_itemsets, metric = "support", min_threshold = 0.4)
rules_table = rules_table[rules_table['confidence']  >= 0.5]
#print(te.columns_)

original_items = te.columns_
items = te.columns_
boughtItems = []
while(True):
    
    if len(original_items) > 0:
        print("Items present in our store(excluding suggestion) : {0} \n".format(items))
    else :
        print("All the items of the store are over.\nThanks for purchasing all the items.\n\n            !!!Visit Again!!!")
        break
    if len(boughtItems) > 0:
        print("You have bought : {0} \n".format(boughtItems))
    print("\n   Choices :- \n1 - For purchasing items.\n2 - For stop purchasing.\n")
    choice = int(input("Enter your choice :- "))
    if choice == 1:
        currentItem = input("\nEnter the Item you want to purchase : ")
        currentList = [currentItem]
        boughtItems.append(currentItem)
        original_items.remove(currentItem)
        try:
            suggestedItem = list(set([list(i)[:] for i in rules_table[rules_table['antecedants'].apply(lambda x: currentItem in list(x))] ['consequents'].tolist()]))
        except IndexError:
            print("There is no more suggestion.")
            items = list(set(original_items) - set(suggestedItem))
        else: 
            if len(suggestedItem) == 0:
                print("There is no more suggestion.")
            else:
                print("\n  !!!Suggestion!!! \nYou can buy {0}:-".format(list(set(suggestedItem) - set(boughtItems))))
            items = list(set(original_items) - set(suggestedItem))
    else :
        print("\nYou have purchased : {0}\n\nThanks for purchasing all the items.\n\n            !!!Visit Again!!!".format(boughtItems))
        break
```
